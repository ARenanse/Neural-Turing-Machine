```
import os
import tensorflow as tf
import NTM_Model
import utilities
import numpy as np
import matplotlib.pyplot as plt

tf.__version__
```

```
batch_size = 60
output_dim = 8
item_len = 3

model = NTM_Model.NTM_Associative_Recall_Model(batch_size,output_dim,item_len,rnn_size=35,memory_rows=128,memory_columns=20,num_read_heads=1, num_write_heads=1)

Cross_Entropy_Loss = NTM_Model.CE_Loss_Function()

optimizer = tf.keras.optimizers.Adam(learning_rate=1 * (1e-4), amsgrad=True,)
```

```
model.load_weights('Associative Model/TRAINING WEIGHTS 3rd Round Controller50/Weight_Start_CEL_End_CEL_12')
```

```
all_training_predictions = []
all_test_predictions = []
logdir = 'Associative Model/MODEL_INFO/MODELGRAPH_2'
writer = tf.summary.create_file_writer(logdir)
tf.summary.trace_on(graph=True, profiler=True)


for i in range(500001):
    num_items = (np.random.randint(2, 6 + 1))
    inputs, labels = utilities.Associative_Recall_InpGen(batch_size, item_len, output_dim, num_items)
    
    print("Epoch: {}".format(i))
    print("Number of items: {}".format(num_items))
    
    
    with tf.GradientTape() as tape:
        
        y_pred, cache = model(inputs)
        loss = Cross_Entropy_Loss(y_true=tf.cast(labels, dtype = tf.float32), y_pred=y_pred)

        if i==0:
          with writer.as_default():
            tf.summary.trace_export(name="ModelGraph",step=0,profiler_outdir=logdir)

        print("Loss at epoch {}: {}".format(i, loss))
        print("*********************************")
        

    grads = tape.gradient(loss, model.variables)
    optimizer.apply_gradients(grads_and_vars=zip(grads, model.variables))

    if i%1000 == 0:

          all_training_predictions.append([inputs,y_pred,cache])

          tf.saved_model.save(model,'Associative Model/Model Variants/3rd Round Controller50/Model_Start_CEL_End_CEL_13') 
          model.save_weights('Associative Model/TRAINING WEIGHTS 3rd Round Controller50/Weight_Start_CEL_End_CEL_13')

          #Deliberately testing it on more than what it is trained for

          test_item_length = (np.random.randint(6,19))
          inputs, labels = utilities.Associative_Recall_InpGen(batch_size, item_len, output_dim, test_item_length)
          y_pred, cache = model(inputs)
          print('Test Sequence Length: ', test_item_length)
          test_loss = Cross_Entropy_Loss(y_true=tf.cast(labels, dtype=tf.float32), y_pred=y_pred)
          print('Test Loss CEL: ', test_loss)
          all_test_predictions.append([inputs, y_pred, cache])

          np.save('Associative Model/Controller50/Start CEL to end CEL 13/ALL_TEST_PREDICTIONS' ,np.array(all_test_predictions))
          np.save('Associative Model/Controller50/Start CEL to end CEL 13/ALL_TRAINING_PREDICTIONS', np.array(all_training_predictions) )
```
